# 연관규칙

```python
# 데이터 구성:Series, DataFrame
import pandas as pd
# 행렬 연산
import numpy as np
# 데이터 전처리 : 항목 값에 대한 index 생성
from mlxtend.preprocessing import TransactionEncoder
# 지지도 계산
from mlxtend.frequent_patterns import apriori
# 연관 규칙
from mlxtend.frequent_patterns import association_rules
```

```python
kurly = pd.read_csv("df_cluster.csv")
kurly.head()
```

```python
kurly.columns
```

```python
# 연관분석을 위한 df
kurly_0 = kurly.loc[kurly['Cluster']==0,['idOrder','ItemSmallName']]
kurly_1 = kurly.loc[kurly['Cluster']==1,['idOrder','ItemSmallName']]
kurly_2 = kurly.loc[kurly['Cluster']==2,['idOrder','ItemSmallName']]


kurly_2.shape
```

## Cluster 0

```python
# 고객id 리스트 생성
id = list(set(kurly_0['idOrder']))
id.sort()

# 고객id별 상품 매칭
list_asso = []
for i in id:
    tmp_list = list(kurly_0[kurly_0['idOrder']==i]['ItemSmallName'])
    tmp_list.sort()
    list_asso.append(tmp_list)
    
list_asso[:1]
```

```python
# item 기준을 포함 여부에 따라 True/False 반환
enc = TransactionEncoder()
kurly_enc = enc.fit_transform(X=list_asso)
df_asso = pd.DataFrame(kurly_enc, columns = enc.columns_)
df_asso
```

```python
# 규제 선택 최소 지지도
v_min_support = 0.5

df_freq = apriori(df_asso, min_support = 0.01, verbose=-1, use_colnames=True)
df_freq.sort_values(by='support', ascending=False).round(3)
```

```python
df_asso_rule_0 = association_rules(df_freq, metric='lift', min_threshold=1)
df_asso_rule_0
```

```python
#### 결과물 정제하기 ####
# dataframe으로 저장
df_asso_rule_0.reset_index(inplace=True, drop=True)
# csv 파일로 결과 저장
df_asso_rule_0.to_csv("cluster_0_association.csv", index=False, encoding='cp949')
# dataframe으로 불러오기
df_0 = pd.read_csv("cluster_0_association.csv", encoding='cp949')
# 특수문자 제거
df_0['antecedents'] = df_0['antecedents'].str[11:-2].str.replace("'", "")
df_0['consequents'] = df_0['consequents'].str[11:-2].str.replace("'", "")
df_0
```

```python
# 다시 csv파일로 저장
df_0.to_csv("cluster_0_association.csv", index=False, encoding='cp949')
```
